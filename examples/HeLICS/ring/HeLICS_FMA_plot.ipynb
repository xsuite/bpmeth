```python
import matplotlib.pyplot as plt
import xtrack as xt
import scipy as sc
import bpmeth
import sympy as sp
import math
import numpy as np
from cpymad.madx import Madx
from sympy.abc import alpha, beta  # Cool notebook greek letters!
import sys
import warnings
import nafflib
```

```python
def makeplot(filename):
    X, Y, Qx, Qy, dQ = np.loadtxt(filename, unpack=True)

    fig, ax = plt.subplots(1, 2, figsize=(9, 3.5))

    sc1 = ax[0].scatter(
        Qx, Qy,
        c=dQ,
        s=0.5,
        cmap="jet",
        rasterized=True
    )

    sc2 = plt.scatter(
        X, Y,
        c=dQ,
        s=0.5,
        cmap="jet",
        rasterized=True
    )

    plt.xlabel("initial x")
    plt.ylabel("initial y")

    ax[0].set_xlabel(r"$Q_x$")
    ax[0].set_ylabel(r"$Q_y$")

    ax[1].set_xlabel(r"initial $x$ [m]")
    ax[1].set_ylabel(r"initial $y$ [m]")

    ax[0].set_xlim(0.33, 0.45)
    ax[0].set_ylim(0.15, 0.27)

    fig.colorbar(sc1, label=r"$\log_{10}(\sqrt{\Delta Q_x^2 + \Delta Q_y^2})$", ax=ax[0])
    fig.colorbar(sc2, label=r"$\log_{10}(\sqrt{\Delta Q_x^2 + \Delta Q_y^2})$", ax=ax[1])
    fig.tight_layout()
    plt.show()
```

```python
makeplot(f"FMA_fullfringe_10000turns.txt")
```
